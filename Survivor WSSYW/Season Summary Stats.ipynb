```python
season_rank_df = pd.read_csv("./data/season_rankings.csv")
word_rank_df = pd.read_csv("./data/ranked_words.csv")
season_sentiment_df = pd.read_csv("./data/season_comments_sentiment.csv")
```

```python
## Plot theme

# Create graphic
# Define colors
color_background = "white"
color_text = "#22211d"

my_theme = (

    # Begin construction of chart
    theme_bw(base_size=15) +

    # Format background colors
    theme(panel_background = element_rect(fill=color_background, color=color_background)) +
    theme(plot_background  = element_rect(fill=color_background, color=color_background)) +
    theme(panel_border     = element_rect(color=color_background)) +
    theme(strip_background = element_rect(fill=color_background, color=color_background)) +

    # Format the grid
    theme(panel_grid_major_y = element_blank()) +
    theme(panel_grid_minor_y = element_blank()) +
    theme(axis_ticks         = element_blank()) +

    # Format the legend
    theme(legend_position = "none") +

    # Format title and axis labels
    theme(plot_title       = element_text(color=color_text, size=12, weight = "bold")) +
    theme(axis_title_x     = element_text(size=10, color="black", weight = "bold")) +
    theme(axis_title_y     = element_text(size=10, color="black", weight = "bold")) +
    theme(axis_text_x      = element_text(size=10, color = color_text)) +
    theme(axis_text_y      = element_text(size=10, vjust=1, color = color_text)) +
    theme(strip_text       = element_text(face = "bold")) 

    # Plot margins
    )
```

```python
def polarity_graph(df):
    mean = df.polarity.quantile(.5)
    q1 = df.polarity.quantile(.25)
    q3 = df.polarity.quantile(.75)
    p = (ggplot(aes(x='polarity', y='score', color='polarity'), df) 
         + geom_point() + geom_rug(sides="b") 
         + geom_vline(aes(xintercept = mean, color='polarity'), linetype='dashed', size=0.5) 
#          + geom_vline(aes(xintercept = q1, color='polarity')) 
#          + geom_vline(aes(xintercept = q3, color='polarity')) 
         + xlim(-1, 1) 
         + scale_color_gradientn(limits=[-1,1], colors=["#F4511E","#FFC107","#B3B6B7","#8BC34A","#4CAF50"]) 
         + scale_y_log10()
         + coord_flip()
        )
    return (p + my_theme + theme(figure_size = (4,6)) + labs(x = "(-)                          Polarity/Sentiment                          (+)") )

print(polarity_graph(season_sentiment_df))
```

```python
# Create the base graph figure. Format the points, lines, axes, set limits, and assign color mapping
cmap_era = {"Classic Era":"#F70020","Golden Age":"#1A7D00","Meta Era":"#0C96F2","Revival Era":"#FB9701","Transitional Era":"#636666","Dark Ages":"#87603E","None":"#BFBEBB"}

def base_fig(df, rank_col, color_col, cmap, show_top=38, show_bottom=1, min_x=1, max_x=10):
    return (
    ggplot(data=df, mapping=aes(x='version', y=rank_col, group='season')) +
    
    # Format data points
    geom_line(aes(color = color_col, alpha = 1), size = 2) +
    geom_point(aes(color = color_col, alpha = 1), size = 4) +
    geom_point(color = "#FFFFFF", size = 1) +
    geom_point(data=df.loc[df['new']==True], mapping=aes(color=color_col, alpha=1), shape='*', size=9) +
    geom_point(data=df.loc[df['new']==True], color='#FFFFFF', shape='*', size=3) 
        +
    
    # Scale reverse for showing top rankings
    scale_y_reverse(limits=[show_bottom,show_top], breaks=range(show_top,show_bottom-1,-1)) +
        scale_x_continuous(limits=[min_x,max_x], breaks=range(min_x-1,max_x+1,1)) +
    
    # Ranking at WSSYW 1.0
#     geom_text(data = df.loc[df['version']=='1.0'],
#         mapping=aes(label = 'season', x = 0.8, y = rank_col, color=color_col), ha = 'right', size=9) +
    # Ranking at WSSYW 9.0
#     geom_text(data = df.loc[df['version']=='9.0'],
#         mapping=aes(label = 'season', x = 9.2, y = rank_col, color=color_col), ha = 'left', size=9) +
    
    # Set limits on graph
#     ylim(show_top,1)  +
    
    # Map colors by color_col value
    scale_color_manual(cmap)
)
```

```python
# Create graph of top 15 seasons by ranking over time
def get_rank_graph(df):
    minRank = int(df['rank'].min())
    maxRank = int(df['rank'].max())
    avgRank = int(df['rank'].mean())
    minX = int(df['version'].min())
    maxX = int(df['version'].max())
    fig = base_fig(df=df, rank_col='rank', color_col='era', cmap=cmap_era, show_top=(maxRank), show_bottom=(minRank), min_x=minX, max_x=maxX)

    top_fig = (fig + labs(x = "r/Survivor WSSYW Version",
            y = "Season Ranking") 
               + annotate(geom="label", x=minX, y=avgRank, label=f"Average Rank: {avgRank}",hjust='left', vjust='top')
               + my_theme + theme(figure_size = (8,4)) #+ coord_cartesian(xlim=(0.5,9.5),ylim=(37,2))
              )
    return(top_fig)

df = season_rank_df[(season_rank_df['season_nbr']==10) & (season_rank_df['rank'].isna()==False)]
print(get_rank_graph(df))
df
```

```python
cmap_tone = {"very positive":"#4CAF50","positive":"#8BC34A","neutral":"#B3B6B7","negative":"#FFC107","very negative":"#F4511E"}

def ranked_word_plot(df, cols = 5):
    
    df['x'] = df['rank'].apply(lambda x: x%cols)
    df['y'] = df['rank'].apply(lambda x: x//5 if (x%5)%2==0  else (x//5)+0.5)
    
    return (
        ggplot(data=df, mapping=aes(x='x', y='y')) + 
        geom_point(color="black") + 
        geom_label(data=df, 
                   mapping=aes(label='word', fill='sentiment'), 
                   color='white',
                   size=10,
                   label_padding=0.2)
        + my_theme
        + scale_y_reverse(limits=[5,0], breaks=range(5,-1,-1))
        + xlim(-.5,4.5)
        + theme(figure_size = (8,2))
        + scale_fill_manual(cmap_tone)
        + theme(axis_title_x=element_blank())
        + theme(axis_title_y=element_blank())
        + theme(axis_text_x=element_blank())
        + theme(axis_text_y=element_blank())
        + labs(title = "Characteristic Words/Phrases") 
    )
    

print(ranked_word_plot(word_rank_df[word_rank_df['season']==10]))
```
